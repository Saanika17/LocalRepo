```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
```

```python
# ---------- MNIST MLP ----------
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
x_train = x_train.reshape((-1,28*28)).astype('float32')/255.0
x_test = x_test.reshape((-1,28*28)).astype('float32')/255.0
```

```python
def build_mlp(input_dim, hidden_layers=[512,256,128], dropout=0.2):

    inp = keras.Input(shape=(input_dim,))
    x = inp
    for h in hidden_layers:
        x = layers.Dense(h, activation='relu')(x)
        x = layers.Dropout(dropout)(x)
    out = layers.Dense(10, activation='softmax')(x)
    model = keras.Model(inp, out)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
mnist_model = build_mlp(28*28)
mnist_model.fit(x_train, y_train, validation_split=0.1, epochs=5, batch_size=128, verbose=2)
print("MNIST eval:", mnist_model.evaluate(x_test, y_test, verbose=0))
```

```python
# ---------- CIFAR-10 MLP ----------
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
x_train = x_train.reshape((-1,32*32*3)).astype('float32')/255.0
x_test = x_test.reshape((-1,32*32*3)).astype('float32')/255.0
y_train = y_train.ravel(); y_test = y_test.ravel()
```

```python
cifar_model = build_mlp(32*32*3, hidden_layers=[1024,512,256], dropout=0.3)
cifar_model.fit(x_train[:20000], y_train[:20000], validation_split=0.1, epochs=8, batch_size=256, verbose=2)
print("CIFAR-10 eval (subset):", cifar_model.evaluate(x_test[:5000], y_test[:5000], verbose=0))
```
